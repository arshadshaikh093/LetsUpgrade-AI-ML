##Range Function
Range is a function that is used to create list of range of numbers.
It will come handy when you want to do some experiments and you don't have a list already. In that case, you can directly use this funciton and you will get a list with range of numbers created for you.

###Parameters
Range function takes three parameters which are as follows:
Start Value: Specifies the start value from where you want yourlist to start
End Value: Specifies the end value from where you want yourlist to start
Step Value: Specifies the difference between two consecutive numbers that you would like to want in your list. By default this value is set to 1.

Usually we use rnage function to iterate over the list and perform some operation or logic on it. Below exmples demonstrates how we can print all the numbers from the range of 10 numbers incremented by 2.


```
for i in range(1,20,2):
  print(i)
```
