```
""" question 1: Write a program to subtract two complex numbers in Python. """

a = complex(5, 3)
b = complex(4,1)
c = a - b
print("subtraction of ", a, " and ", b, " is ", c)
```

```
""" Question 2: Write a program to find the fourth root of a number. """

a = 16
b = a**(1/4)
print(b)
```

```
""" Question 3: Write a program to swap two numbers in Python with the help of a temporary variable. """

a = 5
b = 4
print("Before Swapping: a = ", a, " b = ",b)
c = b
b = a
a = c
print("After Swapping: a = ", a, " b = ",b)
```

```
""" question 4: Write a program to swap two numbers in Python without using a temporary variable. """

a = 5
b = 4
print("Before Swapping: a = ", a, " b = ",b)
a = a+b
b = a-b
a = a-b
print("After Swapping: a = ", a, " b = ",b)
```

```
""" Question 5: Write a program to convert fahrenheit to kelvin and celsius both. """

f = 101
c = (f - 32)/1.80
k = 5/9*(f - 32) + 273.15
print("Farenhiet = ", f)
print("Celsius = ", c)
print("Kelvin = ", k)
```

```
""" Question 6: Write a program to demonstrate all the available data types in Python. Hint: Use type() function. """

integerType = 10
floattype = 10.0
stringType = 'Arshad shaikh'
booleanType = True
listType = [1,2,3,4,5]
tupleType = (1,'Arshad',99.99,True)

print("10 is", type(integerType))
print("10.0", type(floattype))
print("'Arshad shaikh'", type(stringType))
print("True is ", type(booleanType))
print("[1,2,3,4,5] is", type(listType))
print("(1,'Arshad',99.99,True) is ", type(tupleType))
```
