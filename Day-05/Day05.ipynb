```
""" Question 1: Write a Python program to find the first 20 non-even prime natural numbers. """

def isPrimeNumber(num):
  for i in range(2,num):
       if (num % i) == 0:
          break
  else:
    return True  

count = 0
for j in range(3,1000):
  if isPrimeNumber(j):
    print(j)
    count += 1

  if count == 20:
    break  
  
```

```
""" Question 3: Write a Python program to check if the given string is a Palindrome or Anagram or None of them.
Display the message accordingly to the user. """

def isPalindrome(stringvalue):
  reverseString = stringvalue[::-1]
  return stringvalue.lower() == reverseString.lower()
 

def isAnagram(stringValue1, stringValue2):
  return sorted(stringValue1.lower()) == sorted(stringValue2.lower())


print("Ignoring cases for this Palindrome program and hence converting everything to lowercase before comparison")
s1 = input("Enter String1 to check Plaindrome: ")
s2 = input("Enter String2 to check Anagram: ")

if isPalindrome(s1):
    print("Given String ", s1, " is a Plaindrome")
elif isAnagram(s1,s2):
    print("Given Strings ", s1, " and ", s2, " are Anagram of each other")
else:
    print("Given Strings ", s1, " and ", s2, " are None of them")  
```
